```python
import pandas as pd
historical_data = pd.read_csv('history_data.csv')
historical_data.info() 
```

```python
historical_data = historical_data.rename({"Background of Architect": 'background', 
                        "Structure Type": "structure",
                        "Required Construction Materials": "materials",
                        "Characterization of Blueprints": "blueprints",
                        "Is Completed Structure Impossible?": "impossible",
                        "Cost of Structure": "cost"}, axis=1)
```

```python
historical_data["impossible_num"] = [1 if i == "Yes" else 0 for i in historical_data["impossible"]]
```

```python
historical_data.info()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

```

```python
materials = pd.get_dummies(historical_data.materials, prefix='material')
structure = pd.get_dummies(historical_data.structure, prefix='structure')
background = pd.get_dummies(historical_data.background, prefix='background')
blueprints = pd.get_dummies(historical_data.blueprints, prefix='blueprints')
```

```python
data = historical_data
data = data.join(materials)
data = data.join(structure)
data = data.join(background)
data = data.join(blueprints)
```

```python

data.drop(['materials', 'structure', 'background', 'blueprints'], axis=1, inplace=True)
data.head()
```

```python
y = data.impossible_num.copy()
X = data.drop(['cost', 'impossible', 'impossible_num'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
model = LogisticRegression()
model.fit(X, y)
```
